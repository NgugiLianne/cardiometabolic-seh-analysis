**Handling of Missing Data**

The extent of missing data varied substantially across variables. Variables were therefore classified based on the proportion of missing observations to guide subsequent analytical decisions.

Variables with very high levels of missingness (≥60–70%) were considered to have limited analytical utility and were not prioritised for inclusion in multivariable models, as their inclusion could substantially reduce effective sample size and compromise statistical power.

Variables with moderate missingness (approximately 10–40%) were retained for further consideration, as missingness at this level may be addressed through appropriate analytical strategies, including complete-case analysis or imputation, depending on the variable’s role in the analysis and its assumed missingness mechanism.

Variables with low levels of missingness (<10%) were considered suitable for inclusion with minimal impact on sample size and were addressed using standard data-cleaning procedures.

All decisions regarding variable inclusion or exclusion were informed by both the degree of missingness and the substantive relevance of the variable to the study’s research objectives.

## Data Type Harmonization and Structural Validation

Before proceeding to outlier detection and statistical modeling, the dataset underwent a systematic data type harmonization and structural validation process. This step was necessary to ensure that all variables were stored in formats consistent with their conceptual definitions and analytic roles.

Specifically, variables expected to be numeric (e.g., age, anthropometric measures, blood pressure, glucose, and lipid biomarkers) were explicitly coerced to numeric types. Any non-numeric entries arising from data entry artifacts (such as text placeholders or formatting inconsistencies) were converted to missing values and documented. Where this coercion introduced additional missing values, limited re-imputation using median values was applied to preserve analytical consistency, excluding outcome variables by design.

Categorical variables were classified according to their measurement scale (binary, nominal, or ordinal) to ensure correct downstream handling in regression analyses. Ordinal variables, such as wealth quintile, were explicitly defined with their expected ordering.

Finally, the cleaned dataset was structurally compared to the original raw dataset to confirm that no variables were unintentionally dropped or altered during cleaning. Any newly created variables or excluded variables were documented to maintain transparency and reproducibility.

This process ensured that the final analytical dataset was internally consistent, methodologically defensible, and suitable for subsequent outlier detection and inferential analyses.


```python
import pandas as pd
from sklearn.impute import SimpleImputer

# ---------------------------------------------------
# 1. Declare expected variable types
# ---------------------------------------------------

numeric_expected = [
    "Age",
    "bmi","bmi_c","height_cm","weight_kg","waist_cm","hip_cm","waist_hip_ratio",
    "systolic_blood_pressure","diastolic_blood_pressure","blood_glucose",
    "total_cholestrol","ldl_mmol","hdl_mmol","triglycerides",
    "he021","he022","he023","he024","he025","he026"
]

binary_yes_no = [
    "hypertension_status","diabetes_status","cvd_status",
    "pa019","hu001","smoking_ever","alcohol_ever"
]

categorical_nominal = [
    "Sex","education_status","employment_status",
    "marital_status","hiv_status_slf_rpt","smoking_status"
]

ordinal_expected = {
    "wealth_quintile": [1, 2, 3, 4, 5]
}

# ---------------------------------------------------
# 2. Coerce numeric variables safely and audit changes
# ---------------------------------------------------

coerce_report = []

for col in numeric_expected:
    if col in df_clean.columns:
        was_numeric = pd.api.types.is_numeric_dtype(df_clean[col])
        converted = pd.to_numeric(df_clean[col], errors="coerce")

        # Count newly created NaNs
        new_nans = converted.isna().sum() - df_clean[col].isna().sum()

        df_clean[col] = converted
        coerce_report.append((col, was_numeric, new_nans))

coerce_df = pd.DataFrame(
    coerce_report,
    columns=["column", "was_numeric_before", "new_NaNs_created"]
)

coerce_df
```

```python

def extract_numeric_code(series: pd.Series) -> pd.Series:
    """
    Extract leading numeric code from strings like '1 (YES) Yes' → 1, '2 (NO) No' → 2.
    Leaves NaN if no code is found.
    """
    s = series.astype(str).str.extract(r'(\d+)')[0]
    return pd.to_numeric(s, errors='coerce')

def yesno_code_to_bin(series: pd.Series) -> pd.Series:
    """
    Convert Yes/No style coded columns into 0/1.
    Rules:
      - 1 → 1 (Yes)
      - 2 → 0 (No)
      - 'Yes'/'No' strings handled as well
      - Any remaining NaNs imputed to the mode (most frequent)
    Returns int Series (0/1).
    """
    s = series.copy()

    # If it's string-coded, normalize first
    s_str = s.astype(str).str.strip().str.lower()

    # Try numeric extraction (works for '1 (YES) Yes' etc.)
    s_num = pd.to_numeric(s_str, errors="coerce")
    out = pd.Series(np.nan, index=s.index, dtype="float")

    # numeric rules
    out[s_num == 1] = 1
    out[s_num == 2] = 0
    out[s_num == 0] = 0

    # fallback string rules
    map_yes = {"yes", "y", "true", "t", "present", "positive"}
    map_no  = {"no", "n", "false", "f", "absent", "negative"}

    mask_unfilled = out.isna()
    s_unfilled = s_str[mask_unfilled]

    out.loc[mask_unfilled & s_unfilled.isin(map_yes)] = 1
    out.loc[mask_unfilled & s_unfilled.isin(map_no)]  = 0

    # final: fill any remaining NaNs with mode
    if out.isna().any():
        mode_val = out.dropna().mode()
        if len(mode_val):
            out = out.fillna(mode_val.iloc[0])
        else:
            out = out.fillna(0)  # very unlikely fallback

    return out.astype(int)

def clean_string_category(series: pd.Series) -> pd.Series:
    """
    Generic string cleaner for categorical labels:
    - strip whitespace
    - collapse internal spaces
    - title-case (e.g., 'female' -> 'Female', ' self employed ' -> 'Self Employed')
    """
    s = series.astype(str).str.strip()
    # collapse multiple spaces to single
    s = s.str.replace(r"\s+", " ", regex=True)
    return s.str.title()


# --- 3.2 Fix mixed-coded Yes/No variables (produce *_bin 0/1 copies) ---

# Columns known to be in the '1 (YES) Yes' / '2 (NO) No' format in your data
mixed_yesno_cols = [c for c in ["pa012", "pa019", "hu001"] if c in df_clean.columns]

# Keep original columns as-is, but create numeric-code and 0/1 versions
for col in mixed_yesno_cols:
    # numeric code (1/2) version (optional, useful for audits)
    df_clean[col + "_code"] = extract_numeric_code(df_clean[col])
    # binary 0/1 modeling copy
    df_clean[col + "_bin"]  = yesno_code_to_bin(df_clean[col])

df_clean[[c for c in df_clean.columns if c.endswith("_code") or c.endswith("_bin")]].head()
```

```python
# ---------------------------------------------------
# Step 3 — Categorical cleaning & binary recoding
# ---------------------------------------------------

# ---------- Helper functions ----------

def extract_numeric_code(series: pd.Series) -> pd.Series:
    """
    Extract leading numeric code from strings like:
    '1 (YES) Yes' → 1, '2 (NO) No' → 2.
    """
    s = series.astype(str).str.extract(r"(\d+)")[0]
    return pd.to_numeric(s, errors="coerce")


def yesno_code_to_bin(series: pd.Series) -> pd.Series:
    """
    Convert Yes/No style coded variables into binary 0/1.

    Rules:
    - 1 → 1 (Yes)
    - 2 → 0 (No)
    - Common Yes/No strings handled
    - Remaining NaNs filled with mode
    """
    s = series.copy()
    s_str = s.astype(str).str.strip().str.lower()

    out = pd.Series(np.nan, index=s.index, dtype="float")

    # numeric rules
    s_num = pd.to_numeric(s_str, errors="coerce")
    out[s_num == 1] = 1
    out[s_num.isin([0, 2])] = 0

    # string fallbacks
    yes_vals = {"yes", "y", "true", "t", "present", "positive"}
    no_vals  = {"no", "n", "false", "f", "absent", "negative"}

    mask = out.isna()
    out.loc[mask & s_str.isin(yes_vals)] = 1
    out.loc[mask & s_str.isin(no_vals)]  = 0

    # fill remaining with mode
    if out.isna().any():
        mode_val = out.dropna().mode()
        out = out.fillna(mode_val.iloc[0] if len(mode_val) else 0)

    return out.astype(int)


def clean_string_category(series: pd.Series) -> pd.Series:
    """
    Clean categorical string labels safely:
    - strip whitespace
    - collapse internal spaces
    - standardize missing-like values → NA
    - title-case labels
    """
    s = series.astype(str).str.strip()
    s = s.str.replace(r"\s+", " ", regex=True)

    # normalize missing-like values
    missing_like = {
        "nan", "none", "missing", "refused",
        "don't know", "dont know", "dk", ""
    }

    s_lower = s.str.lower()
    s = s.where(~s_lower.isin(missing_like), pd.NA)

    return s.str.title()


# ---------- 3.1 Clean standard categorical variables ----------

categorical_vars = [
    "Sex",
    "education_status",
    "employment_status",
    "marital_status",
    "hiv_status_slf_rpt",
    "smoking_status"
]

for col in categorical_vars:
    if col in df_clean.columns:
        df_clean[col] = clean_string_category(df_clean[col])


# ---------- 3.2 Mixed-coded Yes/No variables → *_code and *_bin ----------

mixed_yesno_cols = [c for c in ["pa012", "pa019", "hu001"] if c in df_clean.columns]

for col in mixed_yesno_cols:
    df_clean[col + "_code"] = extract_numeric_code(df_clean[col])
    df_clean[col + "_bin"]  = yesno_code_to_bin(df_clean[col])


# ---------- 3.3 Quick sanity checks ----------

print("\nBinary variable checks:")
for col in [c for c in df_clean.columns if c.endswith("_bin")]:
    print(f"\n{col}")
    print(df_clean[col].value_counts(dropna=False))

print("\nCategorical variable checks:")
for col in categorical_vars:
    if col in df_clean.columns:
        print(f"\n{col}")
        print(df_clean[col].value_counts(dropna=False).head(10))
```

# Step 5 — Outlier & Range Checks (Continuous Variables)

Before analysis, we’ll flag **implausible values** using domain-informed ranges.
We *won’t* change data yet — just **report** potential issues. After you review, we can either:
- keep as-is (if plausible),
- **winsorize** (cap extremes), or
- remove obvious data-entry errors.

**Plausible ranges** (editable):
- Age: 40–120 (HAALSI cohort)
- BMI: 10–60
- SBP: 70–250 mmHg, DBP: 40–150 mmHg
- Glucose: 2–30 mmol/L
- Lipids (mmol/L): Total 2–12, LDL 0.5–8, HDL 0.2–3, Triglycerides 0.2–10
- Anthropometry: Height 120–220 cm, Weight 25–250 kg, Waist/Hip 40–200 cm, WHR 0.4–1.6


```python
import pandas as pd

# ---------------------------------------------------
# Step 5: Outlier Capping (Biological Plausibility)
# ---------------------------------------------------

# OPTIONAL: Standardize column names (recommended)
df_clean.columns = df_clean.columns.str.strip().str.lower()

# Define bounds (also lowercase to match)
bounds_dict = {
    "age": (40, 110),
    "bmi": (10, 60),
    "systolic_blood_pressure": (60, 260),
    "diastolic_blood_pressure": (30, 150),
    "blood_glucose": (2, 30),
    "total_cholestrol": (2, 12),
    "ldl_mmol": (0.5, 8),
    "hdl_mmol": (0.2, 3),
    "triglycerides": (0.2, 10),
    "height_cm": (130, 210),
    "weight_kg": (30, 220),
    "waist_cm": (50, 180),
    "hip_cm": (50, 180),
    "waist_hip_ratio": (0.5, 1.5),
}

# Track capping results
capping_summary = []

for col, (lower, upper) in bounds_dict.items():

    if col in df_clean.columns:

        # Count outliers
        below_count = (df_clean[col] < lower).sum()
        above_count = (df_clean[col] > upper).sum()
        total_capped = int(below_count + above_count)

        # Store summary if any capping occurs
        if total_capped > 0:
            capping_summary.append({
                "Variable": col,
                "LowerBound": lower,
                "UpperBound": upper,
                "BelowBound": int(below_count),
                "AboveBound": int(above_count),
                "TotalCapped": total_capped
            })

        # Apply capping
        df_clean[col] = df_clean[col].clip(lower=lower, upper=upper)

    else:
        print(f"⚠️ Column '{col}' not found in dataset — skipped.")

# Convert report to DataFrame
capping_report = pd.DataFrame(capping_summary)

if not capping_report.empty:
    capping_report = capping_report.sort_values(by="TotalCapped", ascending=False)
    print("\n✅ Outlier capping completed.\n")
    display(capping_report)
else:
    print("\n✅ No values required capping.")

# ---------------------------------------------------
# Summary statistics after capping
# ---------------------------------------------------

# Only include columns that exist
existing_cols = [col for col in bounds_dict.keys() if col in df_clean.columns]

print("\n📊 Summary statistics after capping:\n")
display(df_clean[existing_cols].describe().T)
```

```python


import numpy as np
import pandas as pd

def yesno_to_bin_general(series: pd.Series) -> pd.Series:
    """
    Convert mixed-coded Yes/No values to 0/1.
    Handles: 1/2, '1 (YES) Yes', '2 (NO) No', 'Yes', 'No', etc.
    Rule: 1 or 'yes' -> 1 ; 2 or 'no' -> 0
    Fills remaining NAs with the mode.
    """
    s = series.copy()

    # Normalize to lowercase strings for regex/text checks
    s_str = s.astype(str).str.strip().str.lower()

    # 1) Try to extract a leading number (e.g., '1' from '1 (yes) yes')
    num = pd.to_numeric(s_str.str.extract(r'^\s*(\d+)')[0], errors='coerce')

    # Map numeric codes if present: 1->1, 2->0
    out = num.map({1: 1, 2: 0})

    # 2) Fill remaining from explicit 'yes'/'no' words
    mask_yes = s_str.str.contains(r'\byes\b', na=False)
    mask_no  = s_str.str.contains(r'\bno\b',  na=False)

    # Where still NaN, set yes -> 1
    out = out.where(~out.isna(), mask_yes.astype(float))
    # Where still NaN, set no -> 0
    out = out.where(~out.isna(), (~mask_yes & mask_no).map({True: 0.0, False: np.nan}))

    # 3) Final fill with mode (if any) to ensure clean ints
    if out.isna().any():
        mode_vals = out.dropna().mode()
        if len(mode_vals):
            out = out.fillna(mode_vals.iloc[0])
        else:
            # default to 0 if no mode found (edge case: column entirely missing)
            out = out.fillna(0)

    return out.astype(int)

# Apply to your columns (only if they exist)
for col in ["pa019", "pa012", "hu001"]:
    if col in df_clean.columns:
        df_clean[col + "_bin"] = yesno_to_bin_general(df_clean[col])

# Quick check
for col in ["pa019_bin", "pa012_bin", "hu001_bin"]:
    if col in df_clean.columns:
        print(f"\n{col} value counts:")
        print(df_clean[col].value_counts(dropna=False))
```

```python
# --------------------------------------------
# Ensure consistent column naming
# --------------------------------------------
df.columns = df.columns.str.strip().str.lower()

# --------------------------------------------
# Define categorical variables to summarise
# --------------------------------------------
categorical_vars = [
    'marital_status',
    'education_status',
    'employment_status',
    'wealth_quintile',
    'smoking_status',
    'alcohol_ever',
    'pa019_bin',
    'hypertension_status',
    'diabetes_status',
    'cvd_status'
]

# Keep only those that actually exist
categorical_vars = [v for v in categorical_vars if v in df.columns]

# --------------------------------------------
# Build categorical descriptive dataframe
# --------------------------------------------
cat_list = []

for var in categorical_vars:
    temp = (
        df
        .groupby(['sex', var])
        .size()
        .reset_index(name='N')
    )

    temp['%'] = temp.groupby('sex')['N'].transform(lambda x: x / x.sum() * 100)
    temp['Variable'] = var
    temp = temp.rename(columns={var: 'Category'})

    cat_list.append(temp)

cat_desc = pd.concat(cat_list, ignore_index=True)

# --------------------------------------------
# Create pivot table
# --------------------------------------------
nice_cat = cat_desc.pivot_table(
    index=["Variable", "Category"],
    columns="sex",
    values=["N", "%"]
).reset_index()

# Flatten column names
nice_cat.columns = ["_".join([str(c) for c in col if c]).strip()
                    for col in nice_cat.columns]

print(nice_cat.head(10))
```
